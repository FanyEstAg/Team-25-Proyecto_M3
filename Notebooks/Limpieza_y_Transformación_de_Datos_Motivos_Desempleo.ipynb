# Transformación de datos : Motivos de desempleo

---

En la presente sección se aplicarán una serie de fuunciones con la finalidad de dar respuesta a los planteamientos de las preguntas clave. 


---



```
#Eliminación columnas hombre y motivos indefinidos

df_desempleo_mujer = df_desempleo_renombre.drop(columns=['total_hombre','hombre_conflictos_laborales','hombre_sin_respuesta_exigencias','hombre_condiciones_riesgosas','hombre_acoso','hombre_pension','hombre_aumento_superacion','hombre_independizarse','hombre_retomar_estudios','hombre_embarazo_familia','hombre_un_familiar','total_otros','hombre_otros','mujer_otros','total_indefinido','hombre_indefinido','mujer_indefinido'])
df_desempleo_mujer
```

```
#Obtencion de porcentajes
df_desempleo_mujer['p_total']= pd.Series((df_desempleo_mujer['total_mujer']/df_desempleo_mujer['total'])*100)
df_desempleo_mujer['p_conflictos_laborales']= pd.Series((df_desempleo_mujer['mujer_conflictos_laborales']/df_desempleo_mujer['total_conflictos_laborales'])*100)
df_desempleo_mujer['p_sin_respuesta_exigencias']= pd.Series((df_desempleo_mujer['mujer_sin_respuesta_exigencias']/df_desempleo_mujer['total_sin_respuesta_exigencias'])*100)
df_desempleo_mujer['p_condiciones_riesgosas']= pd.Series((df_desempleo_mujer['mujer_condiciones_riesgosas']/df_desempleo_mujer['total_condiciones_riesgosas'])*100)
df_desempleo_mujer['p_acoso']= pd.Series((df_desempleo_mujer['mujer_acoso']/df_desempleo_mujer['total_acoso'])*100)
df_desempleo_mujer['p_pension']= pd.Series((df_desempleo_mujer['mujer_pension']/df_desempleo_mujer['total_pension'])*100)
df_desempleo_mujer['p_aumento_superacion']= pd.Series((df_desempleo_mujer['mujer_aumento_superacion']/df_desempleo_mujer['total_aumento_superacion'])*100)
df_desempleo_mujer['p_independizarse']= pd.Series((df_desempleo_mujer['mujer_independizarse']/df_desempleo_mujer['total_independizarse'])*100)
df_desempleo_mujer['p_retomar_estudios']= pd.Series((df_desempleo_mujer['mujer_retomar_estudios']/df_desempleo_mujer['total_retomar_estudios'])*100)
df_desempleo_mujer['p_embarazo_familia']= pd.Series((df_desempleo_mujer['mujer_embarazo_familia']/df_desempleo_mujer['total_embarazo_familia'])*100)
df_desempleo_mujer['p_un_familiar']= pd.Series((df_desempleo_mujer['mujer_un_familiar']/df_desempleo_mujer['total_un_familiar'])*100)
```

```
df_desempleo_mujer
```

```
#Eliminación de columnas sobrantes
df_desempleo_porcentajes = df_desempleo_mujer.drop(columns=['total','total_mujer','total_conflictos_laborales','mujer_conflictos_laborales','total_sin_respuesta_exigencias','mujer_sin_respuesta_exigencias','total_condiciones_riesgosas','mujer_condiciones_riesgosas','total_acoso','mujer_acoso','total_pension','mujer_pension','total_aumento_superacion','mujer_aumento_superacion','total_independizarse','mujer_independizarse','total_retomar_estudios','mujer_retomar_estudios','total_embarazo_familia','mujer_embarazo_familia','total_un_familiar','mujer_un_familiar'])
df_desempleo_porcentajes

```

```
#Obtenemos los menores motivos de despido o pérdida de empleo
df_desempleo_porcentajes.min()
```
