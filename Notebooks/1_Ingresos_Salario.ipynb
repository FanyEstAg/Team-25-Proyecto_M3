#Transformación

* Verificación de los tipo de datos 
* Manipulación de strings
* Uso de map
* Generación de columnas 
* Uso de for loop
* Ordenamiento



```
#Sacar año de la cadena
#Manipulaicón de Strings
def año(p):
  return p.split("/")[0] #romper cadenas cuando se encuentre una "/" y elegir el primer trozo

#Sacar trimestre de la cadena
def trimestre(t):
  return t.split("/")[1] #romper cadenas cuando se encuentre una "/" y elegir el trozo siguiente a este
```

```
df_Salarios.insert(1,"Año", list(map(año, df_Salarios.iloc[:,0])))
#elegir la columna "periodo" con todas las filas
#Usar map para extraer todos los años usando una función creada anteriormente
```

```
df_Salarios.insert(2,"Trimestre", list(map(trimestre, df_Salarios.iloc[:,0])))
```

```
#Eliminar columna de Año/Trimestre
df_Salarios=df_Salarios.drop(df_Salarios.columns[0], axis="columns")
```

```
df_Salarios.head() #Revisión
```

```
df_Salarios.isna().sum() #
```

```
#Crear un nuevo dataframe para almacenar las medias 
df_salarios_media = pd.DataFrame(columns=df_Salarios.columns.to_list()) 
df_salarios_media
```

```
df_Salarios.head()
```

```
#Sacar la media por año a partir de los periodos solo de MUJERES
años=[]
for año in range(2010,2021):
  años.append(año)
  a=df_Salarios.loc[:,"Año"]==str(año)#Filtro de dataframe por año
  s=df_Salarios.loc[a] #Extraer dataframe del filtrado 
  prom=pd.Series(s.mean()) #Promediar los datos
  df_salarios_media=df_salarios_media.append(prom, ignore_index=True) #Añadir el promedio de los valores al nuevo dataframe
 
```

```
df_salarios_media
```

```
#Modificar la columna años 
df_salarios_media=df_salarios_media.drop(columns=["Año"]) #Eliminarla
df_salarios_media.insert(1,"Año", años, False) #Añadir nueva

```

```
df_salarios_media=df_salarios_media.drop(["Trimestre"], axis=1) #Eliminar trimestre
```

```
df_salarios_media.head() #Comprobar
```

```
df_salarios_media.dtypes #Verificar tipo de datos
```

```
#Sacar las sumas por fiñas
suma=[]
for i in range(len(df_salarios_media)):
  total=df_salarios_media.iloc[i,1:]
  suma.append(total.sum())
```

```
df_salarios_media
```

```
#Añadir las sumas totales al data frame
df_salarios_media.insert(7,"Total",suma)
```
